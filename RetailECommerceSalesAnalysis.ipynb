```python
# Replace smart quotes in Product Name
df['Product Name'] = df['Product Name'].str.replace('“', '"').str.replace('”', '"')
```

```python
import re

# Remove non-ASCII characters from Product Name
df['Product Name'] = df['Product Name'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))
```

```python
import matplotlib.pyplot as plt

# Group by Product Name and sum the Sales
sales_by_product = df.groupby('Product Name')['Sales'].sum().sort_values()

# Plot
plt.figure(figsize=(10, 8))
sales_by_product.plot(kind='barh', color='skyblue')

plt.title('Total Sales by Product')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Products by Total Sales
import matplotlib.pyplot as plt

# Compute total sales per product
df['Sales'] = df['Quantity'] * df['Sales']
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
top_products.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 10 Products by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
#Analyze Total Sales Over Time (Monthly Trend)
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
plt.figure(figsize=(10,5))
plt.plot(monthly_sales['Sales'])
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# By Category
#Sales by Category / Sub-Category
df.groupby('Category')['Sales'].sum().plot(kind='bar', figsize=(7,4), title='Sales by Category')

# By Sub-Category
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='barh', figsize=(8,6), title='Sales by Sub-Category')
```

```python
#Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', title='Top 10 Products by Sales', figsize=(8,5))
plt.xlabel("Sales")
```

```python
#Profit Analysis
df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10).plot(kind='barh', title='Top 10 Products by Profit', figsize=(8,5))
plt.xlabel("Profit")
```

```python
#Region/State-wise Sales
df.groupby('State')['Sales'].sum().sort_values().plot(kind='barh', figsize=(8,6), title='Sales by State')
plt.xlabel("Sales")
```

```python
#Segment Analysis (Consumer, Corporate, Home Office)

df.groupby('Segment')['Sales'].sum().plot(kind='bar', title='Sales by Segment')
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Shipping Time (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
#Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Only use numeric columns for correlation
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
#Sales Trend Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.plot(kind='line', figsize=(10, 5), marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.grid(True)
plt.show()
```

```python
# Sales By Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
region_sales.plot(kind='barh', color='skyblue')
plt.title('Sales by Region')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.show()
```

```python
#Customer Segment Performance
segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Sales by Customer Segment')
plt.ylabel('')
plt.show()
```

```python
#Category/Sub-Category Analysis
category_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values()
category_sales.plot(kind='barh', color='orange')
plt.title('Sales by Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.show()
```

```python
#Profit vs Discount Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```
